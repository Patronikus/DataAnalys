```python
user_df = df[['tc','art_sp']]
```

```python
user_df
```

```python
user_df = user_df.rename (columns = {'tc':'user_id', 'art_sp':'brand_info'})
```

```python
def split_data (brand_name):
    return brand_name.split (' ') [-1]
```

```python
user_df['brand'] = user_df.brand_info.apply(split_data)
```

```python
user_df
```

```python
user_df.brand_info.apply(lambda x: x.split(' ') [-1])
```

```python
user_purchases = user_df.groupby ('user_id', as_index = False)\
    .agg({'brand':'count'})\
    .rename (columns = {'brand':'purchases'})\
    .query ('purchases >= 5')
#количество покупок для каждого пользователя (для пользователей, количство покупок которых больше 5)
```

```python
users_uniqie_brands = user_df.groupby('user_id', as_index = False) \
    .agg ({'brand': pd.Series.nunique})\
    .rename (columns = {'brand':'uniqie_brands'})
#количество уникальных брендов
```

```python
lovely_brand_purchases_df =user_df.groupby (['user_id', 'brand'], as_index = False) \
    .agg ({'brand_info':'count'}) \
    .sort_values (['user_id','brand_info'], ascending = [False, False])\
    .groupby('user_id') \
    .head(1)\
    .rename (columns = {'brand':'lovely_brand', 'brand_info':'lovely_brand_purchases'})
#бренд с максимальным количеством покупок для каждого пользователя (любимый бренд)
```

```python
loyalty_df = user_purchases.merge(users_uniqie_brands, on = 'user_id')\
    .merge(lovely_brand_purchases_df, on = 'user_id')
```
