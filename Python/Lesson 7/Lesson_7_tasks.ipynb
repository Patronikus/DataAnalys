Функция `np.where` принимает 3 аргумента - логическую серию, значение для `True` и значение для `False`. 

В результате получается массив (структура, похожая на питоновские списки и пандасовские серии), в котором `True` из логической серии заменены на 2-ой аргумент `np.where`, а `False` - на третий. 

Например,
np.where(df.column1, 'yes', 'no')

```python
chrome_robots = pd.Series(np.where((df.user_type == 'Robots') & (df.browser.str.contains('Chrome')) , 1, 0))
```
