```python
ads_data = df.groupby(['ad_id','event'],as_index = False)\
    .agg ({'user_id':'count'})
```

```python
ads_data = ads_data.pivot (index = 'ad_id', columns = 'event', values = 'user_id').reset_index().fillna(0)
```

```python
ads_data = ads_data.assign(ctr = ads_data.click /ads_data.view )
```

```python
top_ctr = ads_data.query ('click > 0 and view > 0').sort_values('ctr', ascending = False).head(10)
```

```python
top_ctr.to_csv('top_ctr_csv_data.csv', index = False)
```
